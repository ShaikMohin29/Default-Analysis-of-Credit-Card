### Statistical Test

### HYPOTHESIS TESTING:

To check the correlation between categorical and target feature

#### Hypothesis Statements:
$H_{0}$ : There is no dependence between categorical and target feature.

$H_{a}$ : There is a dependency between categorical and target feature.

```python
catg=df1[["SEX","EDUCATION","MARRIAGE"]]
```

```python
from scipy.stats import chi2_contingency
```

```python
for i in catg:
    cont_table = pd.crosstab(df1['default payment next month'],catg[i],margins = True)
    stat,pval,dof,expected=chi2_contingency(cont_table)
    if pval < 0.05:
        print('Feature {} is significant and the pvalue = {} '.format(i,pval))
        
    else:
        print('Feature {} is not significant and the pvalue = {} '.format(i,pval))
```

### HYPOTHESIS TESTING:

To check the correlation between numerical and target feature

#### Hypothesis Statements:
$H_{0}$ : There is no dependence between numerical and target feature.

$H_{a}$ : There is a dependency between numerical and target feature.

```python
import scipy.stats as st
```

```python
num=df1[["LIMIT_BAL","AGE","BILL_AMT1","BILL_AMT2","BILL_AMT3","BILL_AMT4","BILL_AMT5","BILL_AMT6",
        "PAY_AMT1","PAY_AMT2","PAY_AMT3","PAY_AMT4","PAY_AMT5","PAY_AMT6"]]
```

```python
for i in num:
    mi=df1[i][df1["default payment next month"]==0]
    fi=df1[i][df1["default payment next month"]==1]
    stat,pval=st.ttest_ind(mi,fi)
    if pval < 0.05:
        print('Feature {} is significant and the pvalue = {} '.format(i,pval))
        
    else:
        print('Feature {} is not significant and the pvalue = {} '.format(i,pval))
```

### Statistical Test

To check the correlation between categorical and target feature

#### Hypothesis Statements:
$H_{0}$ : There is no dependence between categorical and target feature.

$H_{a}$ : There is a dependency between categorical and target feature..

```python
catg=df1[["SEX","EDUCATION","MARRIAGE",'Cred_Lim1_adv', 'Cred_Lim2_adv',
       'Cred_Lim3_adv', 'Cred_Lim4_adv', 'Cred_Lim5_adv', 'Cred_Lim6_adv',
       'Age_Code']]
from scipy.stats import chi2_contingency
for i in catg:
    cont_table = pd.crosstab(df1['default payment next month'],catg[i],margins = True)
    stat,pval,dof,expected=chi2_contingency(cont_table)
    if pval < 0.05:
        print('Feature {} is significant and the pvalue = {} '.format(i,pval))
    else:
        print('Feature {} is not significant and the pvalue = {} '.format(i,pval))
```

### HYPOTHESIS TESTING:
